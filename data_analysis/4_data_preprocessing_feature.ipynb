# 최종 데이터 변수 추가



```python
import pandas as pd
import numpy as np
import os
```

```python
raw_data = pd.read_csv('data/최종데이터/final_data.csv', encoding='utf-8')
raw_data = raw_data.iloc[:, 1:]
```

```python
# 날짜 형 변환
raw_data['날짜'] = pd.to_datetime(raw_data['날짜'], format='%Y-%m-%d')
```

## 1. 계절 변수 추가
* 봄 : 3 ~ 5월
* 여름 :  6 ~ 8월
* 가을 : 9 ~ 11월
* 겨울 : 12 ~ 2월

```python
season_df = raw_data.copy()
season_df['계절'] = '겨울'
```

```python

# 봄 계절 변경
season_df.loc[((season_df['날짜'].dt.month >=3) & (season_df['날짜'].dt.month <= 5)), '계절'] = '봄'

# 여름
season_df.loc[((season_df['날짜'].dt.month >=6) & (season_df['날짜'].dt.month <= 8)), '계절'] = '여름'

# 가을
season_df.loc[((season_df['날짜'].dt.month >=9) & (season_df['날짜'].dt.month <= 11)), '계절'] = '가을'
```

```python
season_df
```

## 2. 요일 변수 추가
* 각 일자의 요일 변수 추가

```python


season_df['요일'] = season_df['날짜'].dt.dayofweek
season_df
```

```python
weekday = {0:'월', 1:'화', 2:'수', 3:'목', 4:'금', 5:'토', 6:'일'}

week_df = season_df.replace({'요일':weekday})
```

```python
week_df
```

## 3. 기념일 변수 추가

* 복날, 국가대표 국제 경기, 블랙데이

```python
anniv = np.array(['2019-07-22', '2019-08-11', '2019-10-10', '2019-11-14', '2019-11-19', '2019-12-11', '2019-12-15', '2019-12-18', 
 '2020-04-14','2020-07-16', '2020-07-26', '2020-08-15', '2020-01-09', '2020-01-12', '2020-01-15', '2020-01-19', '2020-01-22', '2020-01-26'])

anniv.shape
```

```python
week_df['기념일'] = 0
```

```python
week_df['날짜'] = week_df['날짜'].dt.strftime('%Y-%m-%d')
```

```python
week_df.loc[np.isin(week_df['날짜'], anniv), '기념일'] = 1
```

```python
np.unique(week_df.loc[week_df['기념일']==1, '날짜'])
```

## 4. 공휴일 변수 추가

* 공휴일 전날은 2, 공휴일 당일은 1, 나머지 0
* 황금연휴 포함

```python
from datetime import datetime, timedelta

# 공휴일 날짜
holiday = np.array(['2019-08-15','2019-09-12','2019-09-13','2019-09-14','2019-10-03','2019-10-04','2019-10-05','2019-10-06','2019-10-09',
          '2019-12-25','2020-01-01','2020-01-24','2020-01-25','2020-01-26','2020-01-27','2020-03-01','2020-04-15','2020-04-30','2020-05-01','2020-05-02',
          '2020-05-03','2020-05-04','2020-05-05','2020-06-06','2020-08-15','2020-09-30'])


## 공휴일 전날은 2
day_holiday = pd.to_datetime(holiday) - timedelta(days=1)
day_holiday = day_holiday.strftime('%Y-%m-%d')
```

```python
day_holiday
```

```python
week_df['공휴일'] = 0
```

```python
## 공휴일 전날은 2로 설정
week_df.loc[np.isin(week_df['날짜'], day_holiday), '공휴일'] = 2
```

```python
week_df.loc[week_df['공휴일']==2]
```

```python
## 공휴일 당일은 1로 설정
week_df.loc[np.isin(week_df['날짜'], holiday), '공휴일'] = 1
```

```python
np.unique(week_df.loc[week_df['공휴일']==2, '날짜'])
```
