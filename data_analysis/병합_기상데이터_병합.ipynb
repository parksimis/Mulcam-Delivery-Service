```python
import pandas as pd
import numpy as np
import os
import time
```

```python
data = pd.read_csv('../data/raw_data/종관기상데이터.csv', engine='python', index_col=0)
data
```

```python
## 원하는 지역

location = ['동두천', '수원', '양평', '이천', '파주', '서울']
```

```python
# 경기도, 서울 지점만 추출

pre_data = data.loc[np.isin(data['지점명'], location)].reset_index(drop=True)
```

```python
# 경기도 안의 지점은 모두 경기도로 통일

pre_data.loc[pre_data['지점명']!='서울', '지점명'] = '경기도'
```

```python
# 지점 index 삭제

pre_data.drop('지점', axis=1, inplace=True)
```

```python
# 시간 부분 삭제

pre_data['일시'] = pre_data['일시'].astype('str').str[:-6]
```

```python
# 지점, 일시 별로 평균값 산출

final_data = pre_data.groupby(['지점명', '일시']).mean().reset_index()
```

```python
# 3자리수 반올림

weather = np.round(final_data, 3)
```

```python
# 데이터 확인

weather
```

```python
# csv 파일 내보내기

weather.to_csv('../data/가공/기상데이터.csv', encoding='utf-8')
```
