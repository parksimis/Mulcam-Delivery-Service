# API 활용한 데이터 저장

## 1. 보건복지부_코로나19 시·도발생_현황
* [URL](https://www.data.go.kr/iim/api/selectAPIAcountView.do)
* 보건복지부 코로나19 시도 발생 현황 데이터 결합

```python
# 필요 라이브러리 import
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import time
from tqdm.notebook import tqdm
from urllib import parse
```

```python
# 요청변수

URL = 'http://openapi.data.go.kr/openapi/service/rest/Covid19/getCovid19SidoInfStateJson?serviceKey=Z9e4mmJVa6ltNDI9jdrN2yoxURHKx4WUU6bMZo426Nf5TvBihpTWlzlfInAe6YQELn99oVV6owcdhJdAtEevvQ%3D%3D&pageNo=1&numOfRows=10&startCreateDt=20190717&endCreateDt=20200930'


## API 페이지 가져오기
result = requests.get(URL)
bs_obj = BeautifulSoup(result.content, "html.parser")

item_list = bs_obj.find_all('item')

```

```python
## 컬럼명과 각 태그를 딕셔너리 형태로 구성해서 사용

columns_dic = {'사망자수': 'deathcnt', '시도명': 'gubun', '확진자수': 'defcnt',
              '전일대비증감수': 'incdec', '격리해제수': 'isolclearcnt', '격리중환자수': 'isolingcnt',
               '지역발생수': 'localocccnt', '해외유입수': 'overflowcnt', '10만명당발생률': 'qurrate', '기준일시': 'stdday'}


covid_df = pd.DataFrame(columns = columns_dic.values())
```

```python
# 데이터 가져오기

for i, item in tqdm(enumerate(item_list)):
    for column in columns_dic.values():
        # 각 태그의 텍스트 가져오기
        try:
            covid_df.loc[i, column] = item.find(column).text
        except:
            # 태그가 없는 경우 NULL값 부여
            print(f'{i} 번째 {column}')
            covid_df.loc[i, column] = np.nan
```

```python
# 컬럼명 보기 편하게 이름으로 변경
covid_df.columns = columns_dic.keys()
```

```python
covid_df.iloc[4014]
```

```python
## 검역과 합계에 해당하는 열 삭제

covid_final = covid_df.drop(covid_df.loc[(covid_df['시도명'] == '검역')| (covid_df['시도명'] == '합계')].index, axis=0).reset_index()
```

```python
# 이후 처리가 편하도록 기준일시를 %Y-%m-%d 형태로 변경


covid_final['기준일시'] = covid_final['기준일시'].str[:-5].str.replace(' ', '')

covid_final['기준일시'] = covid_final['기준일시'].str.replace('년', '-').str.replace('월', '-')

covid_final['기준일시'] = pd.to_datetime(covid_final['기준일시'])
```

```python
# 데이터 컬럼 정렬

covid_final = covid_final[['기준일시', '시도명', '전일대비증감수', '확진자수', '사망자수', '격리해제수', '격리중환자수', '지역발생수', '해외유입수', '10만명당발생률']]
```
