```python
import sqlite3

def insert_db(table, values):
    with sqlite3.connect('./db.sqlite3') as conn:
        cur = conn.cursor()
        query = "insert into " +str(table) + "(time, area_name, predict) values(?, ?, ?)"
        cur.executemany(query, values)
        conn.commit()
    cur.close()
    conn.close()
```

```python
import os
import fnmatch
import joblib
from sklearn.linear_model import Ridge

area_list = ['서울', '경기도']
model_dic = {'치킨': 'chicken', '돈까스일식': 'ilsik', '분식': 'bunsik','야식': 'yasik', 
             '족발보쌈': 'jokbal', '중식': 'jungsik', '찜탕': 'jimtang',
             '카페디저트': 'cafe', '패스트푸드': 'fastfood', '한식': 'hansik', '기타': 'etc'}


model_base_path = '../data_analysis/model/'

for file_list in os.listdir(model_base_path): 
    
    for area in area_list:
    
        search_name = area+'*.pkl'
        
        if fnmatch.fnmatch(file_list, search_name):
            basic_data = pd.DataFrame(pd.date_range('2021-03-19', '2021-04-20', freq='H'), columns=['날짜'])
            basic_data = basic_data.iloc[:-1]
            basic_data['날짜'] = basic_data['날짜'].dt.strftime('%Y-%m-%d %H:%M')
            basic_data['광역시도명'] = area
            
            file_split = file_list.split('_')

            model_path = model_base_path + file_list
            loaded_model = joblib.load(model_path)

            X_predict = predict_raw.loc[predict_raw['광역시도명']==area].reset_index(drop=True)
            X_predict.drop('광역시도명', axis=1, inplace=True)
            
            preds = loaded_model.predict(X_predict.values)
            basic_data['예측값'] = np.round(np.exp(preds), 0)
            basic_data['예측값'].astype('int')
            
            table_name = 'main_' + model_dic[file_split[1]]
            insert_db(table_name, basic_data.values.tolist())
        
    
```
