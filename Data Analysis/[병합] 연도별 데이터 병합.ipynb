# 데이터 병합
* 각 연도별로 나눠진 데이터를 하나의 데이터로 합치는 작업 수행



## 1. 서울시 일별 평균 대기오염도 정보


```python
import pandas as pd
import numpy as np
import os
```

```python
air_2019 = pd.read_csv('data/서울시 일별 평균 대기오염도 정보_2019.csv', engine='python')
air_2019.columns = ['측정일시', '측정소명', '이산화질소', '오존', '일산화탄소', '아황산가스', '미세먼지', '초미세먼지']


air_2019 = air_2019.loc[air_2019['측정소명'].str.find('구') !=-1].reset_index(drop=True)
```

```python
air_2019
```

```python
air_2020 = pd.read_csv('data/서울시 일별 평균 대기오염도 정보_2020.csv', encoding='utf-8', index_col=0)

air_2020 = air_2020.loc[air_2020['측정소명'].str.find('구') !=-1].reset_index(drop=True)
```

```python
air_2020
```

```python
air_final = pd.concat([air_2019, air_2020], axis=0)
air_final
```

```python
air_final.to_csv('data/병합데이터/서울시 일별 평균 대기오염도 정보.csv', encoding='utf-8')
```
