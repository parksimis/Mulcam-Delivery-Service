## 코로나 확진자 (서울, 경기도) 파일 병합하기

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### 1) 서울시

```python
raw_seoul = pd.read_csv('../data/서울시 코로나19 확진자 현황.csv', engine='python')
raw_seoul
```

```python
## 확진일 포맷 변경
raw_seoul['확진일']= pd.to_datetime(raw_seoul['확진일'].astype('str'), format='%Y-%m-%d')
```

```python
sort_seoul = raw_seoul.sort_values('확진일', ascending=True)
```

```python
grouped_seoul = sort_seoul.groupby('확진일').count()[['연번']].reset_index()
```

```python
grouped_seoul['광역시도명'] = '서울'
```

```python
grouped_seoul.columns = ['확진일자', '확진자수', '광역시도명']
```

```python
grouped_seoul
```

```python
covid_seoul = grouped_seoul.loc[(grouped_seoul['확진일자']>='2021-01-01')&(grouped_seoul['확진일자']<='2021-04-19')]
```

### 2) 경기도

```python
raw_gg = pd.read_excel('../data/2.xlsx')
raw_gg.head()
```

```python
raw_gg['확진일자'] = pd.to_datetime(raw_gg['확진일자'].astype('str'))
```

```python
raw_gg.info()
```

```python
grouped_gg = raw_gg.groupby('확진일자').count()[['연번']].reset_index()
```

```python
grouped_gg['광역시도명'] = '경기도'
```

```python
grouped_gg.columns = ['확진일자', '확진자수', '광역시도명']
```

```python
grouped_gg
```

```python
covid_gg = grouped_gg.loc[(grouped_gg['확진일자'] >= '2021-01-01')&(grouped_gg['확진일자'] <= '2021-04-19')]
```

```python
covid_final = pd.concat([covid_seoul, covid_gg], axis=0)
```

```python
covid_final.sort_values('확진일자', ascending=True)
```

```python
covid_final.to_csv('data/predict/predict_covid.csv', encoding='utf-8')
```
